```python
def calculadora(operador):
  def operacao(a, b):
    return eval(f"{a} {operador} {b}")

  return operacao

soma = calculadora("+")
print(soma(5, 3))

subtracao = calculadora("-")
print(subtracao(10, 4))

multiplicacao = calculadora("*")
print(multiplicacao(6, 7))

divisao = calculadora("/")
print(divisao(20, 4))
```

```python
def fatorial (n):
  if n == 1:
    return 1
  return n * fatorial(n - 1)

resultado = fatorial(5)
print("O fatorial de 5 e:", resultado)
```

```python
def potencia(a, b):
  if b == 0:
    return 1
  return a * potencia(1, b - 1)
potencia(5, 3)
```

```python
a , b, c = 1, 5, 10
eqgrau2(a, b, c)
def eqgrau2(a, b, c):

  delta = (b ** 2 - 4 * a * c)
  if delta < 0:
    print(f"Delta ({delta}) e menor do que zer0!")
    return None
  rdelta = raiz(delta)
  x1 = (- b - rdelta) / (2 * a)
  x2 = (- b + rdelta) / (2 * a)
  return x1, x2
```

```python
def soma_recursiva(n):
  if n == 1:
    return 1
    return n + soma_recursiva(n - 1)

def soma_iterativa(n):
  soma = 0
  for i in range(1, n + 1):
    soma += 1
  return soma

(soma_recursiva(100), soma_iterativa(100))
```

```python
def fibonacci(n):
  if n <= 2:
    return 1
  return fibonacci(n - 1) + fibonacci(n - 2)

[fibonacci(n) for n in range(1, 8)]
```
